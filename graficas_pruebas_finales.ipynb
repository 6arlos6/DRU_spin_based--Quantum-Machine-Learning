# Resultado entrenamiento 1 qubit

```python
path_data_full_1q = "/content/drive/MyDrive/Notebook_tesis_def/DATA/resultados_clasificacion_1q_iris_3_3.xlsx"
```

```python
df_final_1q = pd.read_excel(path_data_full_1q)
df_final_1q.head()
```

```python
for f in df_final_1q["f_loss"].unique():
  df_f_loss = df_final_1q[df_final_1q["f_loss"] == f]
  plt.figure()
  for noise_ in df_f_loss["Noise"].unique():
    df_f_loss_noise = df_f_loss[df_f_loss["Noise"] == noise_]
    plt.plot(df_f_loss_noise["N_layers"], df_f_loss_noise["accuracy_0"],'-o', label = f"noise = {noise_}")
  plt.title(f"f_loss = {f}")
  plt.legend()
  plt.grid()
  plt.show()
```
