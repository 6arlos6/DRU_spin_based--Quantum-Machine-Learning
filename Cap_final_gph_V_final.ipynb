# 1 qubits

```python
# ========================================================
path_data_full = "/content/drive/MyDrive/Notebook_tesis_def/DATA/Save_excel_1q_iris_3_3(18).xlsx"
df_final = pd.read_excel(path_data_full)

df_final.rename(columns={'f_loss': '$f_{loss}$'}, inplace=True)
df_final.rename(columns={'T2': '$T_{2}$'}, inplace=True)
df_final.rename(columns={'Noise': '$\zeta_{AD}$'}, inplace=True)
df_final.rename(columns={'std_noide': '$\sigma$'}, inplace=True)
# Definir el diccionario de mapeo
mapeo = {'Renyi_Divergence_v2': 'Renyi Divergence',
         'Trace_Distance_v3': 'Trace Distance',
         'Von_Neumman_Divergence_v2': 'Von Neumman Divergence',
         'fidelity_cost': 'Fidelity Cost'}

# Aplicar el mapeo a la columna 'categorias'
df_final['$f_{loss}$'] = df_final['$f_{loss}$'].map(mapeo)

palette = {
    value: sns.color_palette('Set1')[i]
    for i, value in enumerate(df_final['$f_{loss}$'].unique())
}

# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%



df_final_v2 = df_final[df_final["$T_{2}$"] == 0.000023]

# Configurar el estilo de seaborn (opcional)
sns.set(style="whitegrid")

# Crear el scatter plot y asignar colores basados en la columna C
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(x="num_layers", y="accuracy_0",
                               hue="$f_{loss}$",
                               size='$\zeta_{AD}$',
                               data=df_final_v2,
                               palette=palette,
                               style = "$\sigma$",
                               sizes=(20, 200))

# Mostrar el plot
# plt.title(f'Accuracy with T2 = {23} u [s]')
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("$N_{L}$", fontsize=16)
plt.ylabel("Accuracy", fontsize=16)
plt.ylim(0, 1.1);plt.legend(fontsize=12,bbox_to_anchor=(1.05, 1), loc='upper left' ); plt.xticks(fontsize=14); plt.yticks(fontsize=14)
plt.show()

# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
df_final_v2 = df_final[df_final["$T_{2}$"] == 0.000006]

# Configurar el estilo de seaborn (opcional)
sns.set(style="whitegrid")

# Crear el scatter plot y asignar colores basados en la columna C
plt.figure(figsize=(8, 6))
# , hue='C', size='TamaÃ±o', data=df, palette='Set1', sizes=(20, 200)
scatter_plot = sns.scatterplot(x="num_layers", y="accuracy_0",
                               hue="$f_{loss}$",size='$\zeta_{AD}$',
                               data=df_final_v2, palette=palette, sizes=(20, 200),
                               style = "$\sigma$")

# Mostrar el plot
# plt.title(f'Accuracy with T2 = {6} u [s]')
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("$N_{L}$", fontsize=16)
plt.ylabel("Accuracy", fontsize=16)
plt.ylim(0, 1.1);plt.legend(fontsize=12,bbox_to_anchor=(1.05, 1), loc='upper left' ); plt.xticks(fontsize=14); plt.yticks(fontsize=14)
plt.show()



# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
df_final_v2 = df_final[df_final["$\sigma$"] == 0]

# Configurar el estilo de seaborn (opcional)
sns.set(style="whitegrid")

# Crear el scatter plot y asignar colores basados en la columna C
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(x="num_layers", y="accuracy_0", hue="$f_{loss}$",
                               size='$\zeta_{AD}$', data=df_final_v2, palette=palette, sizes=(20, 200),
                               style = "$T_{2}$")

# Mostrar el plot
# plt.title(f'Accuracy with $\sigma$ = {0} [T]')
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("$N_{L}$", fontsize=16)
plt.ylabel("Accuracy", fontsize=16)
plt.ylim(0, 1.1);plt.legend(fontsize=12,bbox_to_anchor=(1.05, 1), loc='upper left' ); plt.xticks(fontsize=14); plt.yticks(fontsize=14)
plt.show()



# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
import matplotlib.pyplot as plt
import seaborn as sns

df_final_v2 = df_final[df_final["$\sigma$"] == 0.001]

# Configurar el estilo de seaborn (opcional)
sns.set(style="whitegrid")

# Crear el scatter plot y asignar colores basados en la columna C
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(x="num_layers", y="accuracy_0", hue="$f_{loss}$",
                               size='$\zeta_{AD}$', data=df_final_v2, palette=palette, sizes=(20, 200),style = "$T_{2}$")

# Mostrar el plot
# plt.title(f'Accuracy with $\sigma$ =  0.001 [T]')
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("$N_{L}$", fontsize=16)
plt.ylabel("Accuracy", fontsize=16)
plt.ylim(0, 1.1);plt.legend(fontsize=12,bbox_to_anchor=(1.05, 1), loc='upper left' ); plt.xticks(fontsize=14); plt.yticks(fontsize=14)
plt.show()


# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
import matplotlib.pyplot as plt
import seaborn as sns

df_final_v2 = df_final[df_final["$\sigma$"] == 0.1]

# Configurar el estilo de seaborn (opcional)
sns.set(style="whitegrid")

# Crear el scatter plot y asignar colores basados en la columna C
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(x="num_layers", y="accuracy_0", hue="$f_{loss}$",
                               size='$\zeta_{AD}$', data=df_final_v2, palette=palette, sizes=(20, 200),
                               style = "$T_{2}$")

# Mostrar el plot
# plt.title(f'Accuracy with $\sigma$ =  0.1 [T]')
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("$N_{L}$", fontsize=16)
plt.ylabel("Accuracy", fontsize=16)
plt.ylim(0, 1.1);plt.legend(fontsize=12,bbox_to_anchor=(1.05, 1), loc='upper left' ); plt.xticks(fontsize=14); plt.yticks(fontsize=14)
plt.show()
```
